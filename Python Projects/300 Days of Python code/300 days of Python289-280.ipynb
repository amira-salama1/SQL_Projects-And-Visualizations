```python
#  Define a function  mystery  with arguments(a,b,c) add condition if a > b return a else c
# https://www.clcoding.com/2024/12/python-coding-challenge-day-289-what-is.html
def mystery(a, b, c):
    return a if a > b else c

result = mystery(5,3,10)

print(result)
```

```python
# Modify a Dictionary with range of 3 by adding a number at key 3 = 9:
#https://www.clcoding.com/2024/12/python-coding-challenge-day-288-what-is.html
x = {i:i**2 for i in range(3)}
y = x.keys()
x[3] = 9

print(list(y))
print(y)
print(x)
```

```python
#https://www.clcoding.com/2024/12/python-coding-challenge-day-287-what-is.html
def func(nums):
    result = 1
    for n in nums:
        result *= n
    return result

print(func([1,2,3]))
```

```python
#https://www.clcoding.com/2024/12/python-coding-challenge-day-286-what-is.html
def recursive_sum(n):
    if n == 0:
        return 0
    return n + recursive_sum(n-1)

print(recursive_sum(5))
print(5+4+3+2+1)
```

```python
#https://www.clcoding.com/2024/12/python-coding-challenge-day-285-what-is.html
# argument unpacking
def func(a,b,c):
    print(a,b,c)
args = (1,2,3)

func(*args)
```

```python
#https://www.clcoding.com/2024/12/python-coding-challenge-day-284-what-is.html
class MyClass:
    @classmethod
    def hello(cls):
        print(f"Hello from {cls.__name__}")

MyClass.hello()
```
