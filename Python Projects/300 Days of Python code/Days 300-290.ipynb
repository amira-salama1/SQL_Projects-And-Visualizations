```python
# 5- get the factorial of a number using a function
# The factorial function (symbol: !) says to multiply all whole numbers from our chosen number down to 1.
# example: 4! = 4 × 3 × 2 × 1 = 24
# Source: https://www.mathsisfun.com/numbers/factorial.html


def factorial(n):
    if n == 1 or n == 0 :
        return 1
    else: 
        return n * factorial(n-1)

print(factorial(5))
```

```python
# using the factorial function in Math library
import math
result = math.factorial(5)
print(result)


result= math.factorial(10)
print('the result of factorial 10! is', result)


result = 10 * math.factorial(9)
print('the result of factorial 10! is also the multiplication of the number * the factorial of that number minus1: ', result)
```

```python
# 8- define a function that return the sum of the integer by itself - 1 until it reaches 0

def recursive_sum(n):
    if n == 0 :
        return 0
    else: 
        return n + recursive_sum(n-1)

print(recursive_sum(4))
```
